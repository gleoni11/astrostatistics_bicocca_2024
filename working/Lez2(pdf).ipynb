```python
import numpy as np
import matplotlib.pyplot as plt

x=np.random.uniform(0.1, 10, 1000)

plt.hist(x, bins=50)
plt.xlabel('x')
plt.show()

y=np.log10(x)
plt.hist(y, bins=50, density=True, label="log10(x)") 

x1 = np.linspace(0.1, 10, 1000) 
y1=np.log10(x1)
py= x1*np.log(10)*(1/(10-0.1))
plt.plot(y1,py, color='red', label="teoretical pdf")
plt.legend()
plt.show()

print("log10(mean(x)): ", np.log10(np.mean(x)))
print("mean(y): ", np.mean(y))
print("log10(median(x)): ", np.log10(np.median(x)))
print("median(y)", np.median(y))
```
