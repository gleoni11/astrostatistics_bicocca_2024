```python
import numpy as np
import matplotlib.pyplot as plt

x=np.random.uniform(0.1, 10, 1000)

plt.hist(x, bins=50)
plt.xlabel('x (valori uniformi casuali)')
plt.show()

y=np.log10(x)
plt.hist(y, bins=50, density=True) #density=True->così sto dicendo che l'isto è di una densità di probabilità, quindi in automatico fa l'isto in modo che l'area sottesa sia 1

x1 = np.linspace(0.1, 10, 1000) #crea 1000 punti tra 0.1 e 10 tutti equidistanti tra loro
y1=np.log10(x1)
py= x1*np.log(10)*(1/9.9)
plt.plot(y1,py, color='red')
plt.show()

print(np.log10(np.mean(x)))
print(np.mean(y))
print(np.log10(np.median(x)))
print(np.median(y))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1, 10, 1000) #crea 1000 punti tra 0.1 e 10 tutti equidistanti tra loro
y=np.log10(x)
py= x*np.log(10)*(1/(10-0.1)) #x=10^y 
plt.plot(y,py, color='red')

#la pdf è normalizzata se l'area sotto la curva fa 1!!! Quindi ci sta che l'altezza sia più di 1 

#plt.ylim(0, 80)
#plt.plot(x,y)

plt.show()
```
